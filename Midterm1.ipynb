```python
# importing the required module 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#Optional for disabling warnings 
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
#Set figure resolution to 300dpi
import matplotlib as mpl
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
df = pd.read_excel (r'./Anatomy.xlsx', index_col=0) 
df.describe()
```

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# Plot Histogram on x
x1 = df.percent
plt.hist(x1, bins = 25, edgecolor='k')
plt.gca().set(title='Frequency Histogram - Midterm Exam', ylabel='Frequency');
plt.xlim(0,1)

plt.axvline(x1.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*1.05, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()))
```

```python
kwargs = dict(hist_kws={'alpha':.4}, kde_kws={'linewidth':3})

plt.figure(figsize=(10,7), dpi= 300)
sns.distplot(x1, bins = 25, color="dodgerblue", label="Original", **kwargs)
plt.axvline(x1.mean(), color='dodgerblue', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*1.05, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()), color="dodgerblue")

plt.xlim(0,1)
plt.legend();
plt.gca().set(title='Fall 2023 Bio 132 - Exam 1 Scores', ylabel='Density', xlabel='Score')
```
